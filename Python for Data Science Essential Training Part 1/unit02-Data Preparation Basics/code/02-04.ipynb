## Segment 4 - Concatenating and transforming data

```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

```python
# Let's create a DataFrame 6x6 with values from 0-35
DF_obj = pd.DataFrame(np.arange(36).reshape(6,6))
DF_obj
```

```python
# Let's create another Dataframe 5x3 with values from 0-14
DF_obj_2 = pd.DataFrame(np.arange(15).reshape(5, 3))
DF_obj_2
```

### Concatenating data

```python
# to concat objects based on their row index values, you call pd.concat method on the objects you want joined and pass the axis=1 argument 
# so columns will be added based on the row index values
pd.concat([DF_obj, DF_obj_2], axis=1)
```

```python
# Also we can concat rows based on column index
# in this case we leave out axis=1
pd.concat([DF_obj, DF_obj_2])
```

### Transforming data
#### Dropping data

```python
# we use the `drop` method to drop data from DataFrame
# By default `drop` will drop rows index with specified position
DF_obj.drop([0, 2])
```

```python
# to drop columns we pass axis=1
DF_obj.drop([0, 2], axis=1)
```

### Adding data

```python
# We will create a Series object that we can use to add as a variable to our DataFrame(s)
series_obj = Series(np.arange(6))

# We give our series a name
series_obj.name = "added_variable"
series_obj
```

```python
# Now we add our series object to the dataframe using `join`
variable_added = DataFrame.join(DF_obj, series_obj)
variable_added
```

```python
# instead of `join`, we will use the `append` method
# we will append the `variable_added` to itself and leave the index values the same as they were
# so we will add argument ignore_index = false
added_datatable = variable_added.append(variable_added, ignore_index=False)
added_datatable
```

```python
# to regenerate the index, we set ignore_index=True
added_datatable = variable_added.append(variable_added, ignore_index=True)
added_datatable
```

### Sorting data

```python
# we are going to use `sort_values` method
DF_sorted = DF_obj.sort_values(by=(5), ascending=[False])
DF_sorted
```
